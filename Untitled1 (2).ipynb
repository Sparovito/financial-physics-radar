```python
# @title
# === Selettore da CSV: categorie + ricerca + scelta finale in TICKER ===
from ipywidgets import Dropdown, Text, Button, HBox, VBox, HTML
from IPython.display import display, clear_output
import pandas as pd
import os

CSV_PATH = "instruments.csv"  # cambia se usi un altro nome/percorso

# 1) Carica CSV (auto-detect separatore , o ;)
if not os.path.exists(CSV_PATH):
    raise FileNotFoundError(
        f"Non trovo {CSV_PATH}. Caricalo in Colab (File → Upload) o crea il template con il blocco sopra."
    )

df = pd.read_csv(CSV_PATH, sep=None, engine="python").dropna()
df = df.rename(columns={c:c.lower() for c in df.columns})
assert set(["category","label","ticker"]).issubset(df.columns), \
    "Il CSV deve avere colonne: category,label,ticker"

# (sanity) dedup su ticker mantenendo la prima occorrenza
df = df.drop_duplicates(subset=["ticker"], keep="first")

# 2) Costruisci mapping categoria -> [(etichetta mostrata, valore ticker)]
def build_options(subdf):
    # etichetta “ricca” mostrata a sinistra
    return [(f"{row['label']}  —  {row['ticker']}", row["ticker"]) for _, row in subdf.iterrows()]

categories = sorted(df["category"].unique().tolist())

cat_dd   = Dropdown(options=categories, value=categories[0], description="Categoria:")
search_t = Text(value="", placeholder="Filtra per nome o ticker...")
inst_dd  = Dropdown(options=build_options(df[df["category"]==categories[0]]),
                    value=build_options(df[df["category"]==categories[0]])[0][1],
                    description="Strumento:")
txt_custom = Text(value="", placeholder="...oppure scrivi un ticker manuale (es. NVDA, ^GSPC, GC=F)")
btn_ok   = Button(description="Usa questo", button_style="success")
msg      = HTML("")

def refresh_instruments():
    cat = cat_dd.value
    q = search_t.value.strip().lower()
    sub = df[df["category"]==cat]
    if q:
        sub = sub[(sub["label"].str.lower().str.contains(q)) | (sub["ticker"].str.lower().str.contains(q))]
    opts = build_options(sub)
    if not opts:
        opts = [("— Nessun risultato —", "__NONE__")]
    inst_dd.options = opts
    inst_dd.value   = opts[0][1]

def on_cat_change(change):
    if change["name"]=="value":
        refresh_instruments()

def on_search_change(change):
    if change["name"]=="value":
        refresh_instruments()

cat_dd.observe(on_cat_change, names="value")
search_t.observe(on_search_change, names="value")

def on_confirm(_):
    clear_output(wait=True)
    chosen = txt_custom.value.strip().upper() or inst_dd.value
    if chosen == "__NONE__" or not chosen:
        display(VBox([cat_dd, HBox([search_t]), inst_dd, txt_custom, btn_ok, HTML("<b>Selezione non valida.</b>")]))
        return
    # trova etichetta leggibile corrente (se esiste)
    try:
        lbl = next(lbl for (lbl, val) in inst_dd.options if val == inst_dd.value)
    except StopIteration:
        lbl = chosen
    print(f"✅ Ticker selezionato: {chosen}  (da: {lbl})")
    global TICKER
    TICKER = chosen

btn_ok.on_click(on_confirm)

# primo render
refresh_instruments()
display(VBox([cat_dd, HBox([search_t]), inst_dd, txt_custom, btn_ok, msg]))
```

```python
# @title
# === Selettore INTERVAL/PERIOD per yfinance ===
from ipywidgets import Dropdown, Button, HBox, VBox, HTML
from IPython.display import display, clear_output
import re

# Intervalli supportati da yfinance
interval_options = [
    "1m","2m","5m","15m","30m","60m","90m","1h",
    "1d","5d","1wk","1mo","3mo"
]
# Periodi supportati da yfinance
period_options = [
    "1d","5d","1mo","3mo","6mo","1y","2y","5y","10y","ytd","max"
]

dd_interval = Dropdown(options=interval_options, value="1d", description="INTERVAL:")
dd_period  = Dropdown(options=period_options,  value="5y", description="PERIOD:")

btn_set = Button(description="Imposta", button_style="success")
msg = HTML()

# Stima giorni per validare minute/hourly vs period (yfinance: granularità < 1d ammessa solo fino a ~60 giorni)
def period_to_days(p):
    if p == "ytd":  # approx: dall'inizio anno ad oggi
        return 365  # non preciso, ma ci basta per validazione
    if p == "max":
        return 36500
    m = re.fullmatch(r"(\d+)([dwmy])", p)
    if not m:
        # valori come 1d, 5d, 1mo, 3mo, 6mo, 1y, 2y, 5y, 10y già gestiti sotto, ma fallback:
        return 36500
    n, unit = int(m.group(1)), m.group(2)
    if unit == "d": return n
    if unit == "w": return n*7
    if unit == "m": return n*30
    if unit == "y": return n*365
    return 36500

def approx_days(p):
    # mapping diretto per le stringhe standard
    mapping = {
        "1d":1, "5d":5, "1mo":30, "3mo":90, "6mo":180,
        "1y":365, "2y":730, "5y":1825, "10y":3650,
        "ytd":365, "max":36500
    }
    return mapping.get(p, period_to_days(p))

def on_click(_):
    clear_output(wait=True)
    interval = dd_interval.value
    period   = dd_period.value
    # Validazione yfinance: minuti/ore ammessi solo con period <= ~60 giorni
    intraday = interval in {"1m","2m","5m","15m","30m","60m","90m","1h"}
    days = approx_days(period)
    warn = ""
    if intraday and days > 60:
        warn = ("<br><span style='color:#b45309'>⚠️ Con intervalli intraday "
                "(<code>m/h</code>) Yahoo Finance supporta solo periodi ≤ ~60 giorni. "
                "Considera di scegliere <b>1d</b>/<b>5d</b>/<b>1mo</b>/<b>3mo</b> "
                "oppure passa a un INTERVAL ≥ 1d.</span>")
    print(f"✅ INTERVAL selezionato: {interval} — PERIOD selezionato: {period}")
    if warn: display(HTML(warn))
    global INTERVAL, PERIOD
    INTERVAL = interval
    PERIOD = period

display(VBox([HBox([dd_interval, dd_period]), btn_set, msg]))
btn_set.on_click(on_click)
```

```python
# @title
# Scarico dati
data = yf.download(TICKER, period=PERIOD, interval=INTERVAL, auto_adjust=True, progress=False)
if data.empty:
    raise ValueError("Nessun dato. Cambia TICKER/INTERVAL/PERIOD.")
px = data["Close"].dropna().copy()
px.index = pd.to_datetime(px.index)

# Ritorni log detrendati (fix 1D)
rets = np.log(px).diff().dropna()
x = rets.to_numpy(dtype=float).ravel()
x_dt = signal.detrend(x, type="linear")
rets = pd.Series(x_dt, index=rets.index)

# FFT
N = len(rets)
freqs = np.fft.rfftfreq(N, d=1)
fft_vals = np.fft.rfft(rets.values)
power = np.abs(fft_vals)

mask = freqs > 0
top_idx = np.argsort(power[mask])[::-1][:TOP_K]
top_freqs = np.sort(freqs[mask][top_idx])

plt.figure(figsize=(10,4))
plt.plot(freqs[mask], power[mask])
plt.title(f"{TICKER} – Spettro FFT (magnitudine)")
plt.xlabel("Frequenza (1/barra)"); plt.ylabel("Magnitudine"); plt.grid(True); plt.show()

print("Top frequenze:", np.round(top_freqs, 6))
print("Top periodi (barre):", (1/np.maximum(top_freqs,1e-12)).astype(int))
```

```python
# @title
# === 4) FFT su LOG-PREZZO (low-pass) con vettori 1D robusti ===
import numpy as np, pandas as pd, matplotlib.dates as mdates
import matplotlib.pyplot as plt

# Assicura che px sia una Serie 1D
if isinstance(px, pd.DataFrame):
    px = px.squeeze("columns")
px = pd.Series(px).dropna()
px.index = pd.to_datetime(px.index)

# 1) Log-prezzo (1D)
lp = np.log(px).astype(float)
vals = lp.to_numpy().ravel()           # <-- 1D garantito
N = vals.shape[0]

# 2) FFT (1D) e filtro low-pass
freqs = np.fft.rfftfreq(N, d=1.0)
fft_vals = np.fft.rfft(vals)           # 1D (N//2 + 1,)

CUTOFF_PCT = 0.03                      # prova 0.02–0.05
cut = max(2, int(len(fft_vals) * CUTOFF_PCT))

keep = np.zeros_like(fft_vals, dtype=bool)  # 1D
keep[:cut] = True                       # tiene le freq più basse (trend + cicli lenti)

fft_filt = np.zeros_like(fft_vals, dtype=complex)  # 1D
fft_filt[keep] = fft_vals[keep]

# 3) iRFFT (1D) -> log-prezzo ricostruito
lp_recon = np.fft.irfft(fft_filt, n=N)          # 1D (N,)
lp_recon = pd.Series(lp_recon, index=lp.index)  # indice = date

# 4) Torna ai prezzi
px_recon = np.exp(lp_recon)

# 5) Plot allineato
idx = px.index.intersection(px_recon.index)
fig, ax = plt.subplots(figsize=(10,4))
ax.plot(idx, px.reindex(idx).values, label="Prezzo reale", linewidth=1.4)
ax.plot(idx, px_recon.reindex(idx).values, "--", label=f"FFT low-pass (~{int(CUTOFF_PCT*100)}%)", linewidth=1.6)
loc = mdates.AutoDateLocator(); ax.xaxis.set_major_locator(loc); ax.xaxis.set_major_formatter(mdates.ConciseDateFormatter(loc))
ax.set_title(f"{TICKER} – Prezzo vs Ricostruzione FFT (log-prezzo, low-pass)"); ax.grid(True); ax.legend(); plt.show()

print("Shapes — vals:", vals.shape, "| fft_vals:", fft_vals.shape, "| keep:", keep.shape)
print("N barre:", N, "| cut bins:", cut, "| f_max tenuta:", freqs[cut-1] if cut>0 else 0.0)
```

```python
# @title
# === 5A) Rolling FFT low-pass sul log-prezzo ===
import numpy as np, pandas as pd, matplotlib.pyplot as plt, matplotlib.dates as mdates

lp = np.log(px).astype(float)
W = 256          # ampiezza finestra (⬆︎ = più liscio). Prova 128–512
CUTOFF_PCT = 0.03  # % dei bin più bassi da tenere per finestra

def fft_lowpass_last(vals, cutoff_pct):
    """Restituisce SOLO l'ULTIMO punto filtrato della finestra per ridurre il leakage."""
    n = len(vals)
    fv = np.fft.rfft(vals)
    cut = max(2, int(len(fv)*cutoff_pct))
    mask = np.zeros_like(fv, dtype=bool)
    mask[:cut] = True
    fv_f = np.zeros_like(fv, dtype=complex); fv_f[mask] = fv[mask]
    recon = np.fft.irfft(fv_f, n=n)
    return recon[-1]  # ultimo punto “denoised” della finestra

# costruisco serie filtrata “online”: a ogni t prendo l’ultimo punto della finestra [t-W+1 ... t]
lp_roll = []
idx = []
vals = lp.to_numpy()
for t in range(W-1, len(vals)):
    lp_last = fft_lowpass_last(vals[t-W+1:t+1], CUTOFF_PCT)
    lp_roll.append(lp_last)
    idx.append(lp.index[t])

lp_roll = pd.Series(lp_roll, index=pd.to_datetime(idx))
px_roll = np.exp(lp_roll)  # torna ai prezzi

# Plot
i = px.index.intersection(px_roll.index)
fig, ax = plt.subplots(figsize=(10,4))
ax.plot(px.index, px.values, label="Prezzo reale", linewidth=1.2)
ax.plot(i, px_roll.reindex(i).values, "--", label=f"Rolling FFT low-pass (W={W}, ~{int(CUTOFF_PCT*100)}%)", linewidth=1.6)
loc = mdates.AutoDateLocator(); ax.xaxis.set_major_locator(loc); ax.xaxis.set_major_formatter(mdates.ConciseDateFormatter(loc))
ax.set_title(f"{TICKER} – Rolling FFT low-pass (niente crollo ai bordi)"); ax.grid(True); ax.legend(); plt.show()
```

```python
# @title
# === 5B) Previsione 1-passo con rolling FFT low-pass ===

# allinea px e lp_roll (hanno indici diversi all'inizio)
common = px.index.intersection(lp_roll.index)
lp_f = lp_roll.reindex(common)
px_f = px.reindex(common)

# deriva filtrata (log-return previsto dal filtro): Δlpf_t = lpf_t - lpf_{t-1}
dlpf = lp_f.diff()

# previsione one-step: p_hat_{t+1} = p_t * exp(Δlpf_t)
# per confrontare con realtà, spostiamo di 1 in avanti
p_hat = (px_f * np.exp(dlpf)).shift(1)

# target reale = prezzo successivo
p_real_next = px_f.shift(-1)

# allinea e taglia NaN
bt_idx = p_hat.index.intersection(p_real_next.index)
p_hat = p_hat.reindex(bt_idx)
p_real_next = p_real_next.reindex(bt_idx)

mask = p_hat.notna() & p_real_next.notna()
p_hat = p_hat[mask]; p_real_next = p_real_next[mask]

# metriche: MAE su prezzo, hit-rate direzione su ritorni
ret_hat = np.log(p_hat).diff()
ret_real = np.log(p_real_next).diff()

import numpy as np
mae_price = np.mean(np.abs(p_hat - p_real_next))
hit = np.mean(np.sign(ret_hat.dropna().align(ret_real.dropna(), join="inner")[0]) ==
              np.sign(ret_real.dropna().align(ret_hat.dropna(), join="inner")[0]))

print(f"Backtest NVDA — MAE prezzo: {mae_price:.3f} | Hit-rate direzione: {hit:.3f} (baseline~0.50)")

# plot breve per vedere lag/fit
show_n = 300  # ultimi N punti
ii = p_hat.index[-show_n:]
plt.figure(figsize=(10,4))
plt.plot(ii, p_real_next.reindex(ii).values, label="Prezzo reale (t+1)", linewidth=1.2)
plt.plot(ii, p_hat.reindex(ii).values, "--", label="Previsione (t+1) da Rolling FFT", linewidth=1.2)
plt.title(f"{TICKER} – One-step forecast (ultimi {show_n})")
plt.grid(True); plt.legend(); plt.show()
```

```python
# @title
# === 5C) Evidenzia FUTURO + forecast a H passi ===

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

HORIZON = 30       # <-- cambia qui: 1, 3, 5, ...

# Ricordati: abbiamo già px_f (prezzi allineati) e lp_roll (log-prezzo filtrato rolling)
# dlpf = lp_f.diff() calcolato in 5B. Lo ricalcolo per sicurezza:
common = px.index.intersection(lp_roll.index)
lp_f = lp_roll.reindex(common)
px_f = px.reindex(common)
dlpf = lp_f.diff()

# ---- Previsione a H passi ----
# Baseline semplice: p_hat_h(t) = p(t) * exp( dlpf(t) ), poi shift h in avanti
# (cioè la previsione fatta a t viene confrontata con il valore reale a t+H)
p_hat_h = (px_f * np.exp(dlpf)).shift(HORIZON)     # previsione di t+H
p_real_next_h = px_f.shift(-HORIZON)               # target reale a t+H

# Allineamento & maschere
bt_idx = p_hat_h.index.intersection(p_real_next_h.index)
mask = p_hat_h.notna() & p_real_next_h.notna()
bt_idx = bt_idx[mask.reindex(bt_idx, fill_value=False)]
p_hat_h = p_hat_h.reindex(bt_idx)
p_real_next_h = p_real_next_h.reindex(bt_idx)

# Metriche semplici
mae_h = np.mean(np.abs(p_hat_h - p_real_next_h))
ret_hat_h  = np.log(p_hat_h).diff()
ret_real_h = np.log(p_real_next_h).diff()
hit_h = np.mean(np.sign(ret_hat_h.dropna().align(ret_real_h.dropna(), join="inner")[0]) ==
                np.sign(ret_real_h.dropna().align(ret_hat_h.dropna(), join="inner")[0]))
print(f"H={HORIZON} — MAE: {mae_h:.3f} | Hit-rate: {hit_h:.3f} (baseline ~0.50)")

# ---- Plot ultimi N punti con area FUTURO ----
show_n = 300
ii = bt_idx[-show_n:]

# La “zona futuro” sono gli ultimi HORIZON punti della finestra mostrata
if len(ii) > HORIZON:
    future_start = ii[-HORIZON]
    future_end   = ii[-1]
else:
    future_start = ii[0]; future_end = ii[-1]

fig, ax = plt.subplots(figsize=(10,4))
ax.plot(ii, p_real_next_h.reindex(ii).values, label=f"Prezzo reale (t+{HORIZON})", linewidth=1.3)
ax.plot(ii, p_hat_h.reindex(ii).values, "--", label=f"Previsione (t+{HORIZON}) da Rolling FFT", linewidth=1.3)

# area colorata “futuro”
ax.axvspan(future_start, future_end, alpha=0.12, label="Zona FUTURO (orizzonte)", ymin=0, ymax=1)

loc = mdates.AutoDateLocator()
ax.xaxis.set_major_locator(loc)
ax.xaxis.set_major_formatter(mdates.ConciseDateFormatter(loc))
ax.set_title(f"{TICKER} — Forecast a H={HORIZON} passi + evidenza FUTURO")
ax.grid(True); ax.legend(); plt.show()

# (opzionale) linea verticale al confine passato/futuro
fig, ax = plt.subplots(figsize=(10,3.5))
ax.plot(ii, p_real_next_h.reindex(ii).values, label=f"Reale (t+{HORIZON})", linewidth=1.1)
ax.plot(ii, p_hat_h.reindex(ii).values, "--", label="Prev.", linewidth=1.1)

ax.axvline(future_start, linestyle=":", linewidth=1.2, label="Confine Futuro")
ax.grid(True); ax.legend()
ax.set_title(f"{TICKER} — Confine e previsione a H={HORIZON}")
ax.xaxis.set_major_locator(loc)
ax.xaxis.set_major_formatter(mdates.ConciseDateFormatter(loc))
plt.show()
```

```python
# @title
# === Frequenze dominanti & sintesi scenari dal log-prezzo ===
import numpy as np, pandas as pd, matplotlib.pyplot as plt, matplotlib.dates as mdates

# ----- Parametri -----
TOP_K        = 8        # quante frequenze tenere
HORIZON      = 60       # giorni futuri da generare
N_SCENARIOS  = 5        # quanti scenari sintetici
PHASE_JITTER = 0.8      # deviazione std delle fasi (rad) per scenari "nuovi"
AMP_SCALE    = 1.0      # scala ampiezze (>=0). Es: 1.2 = +20% oscillazione

# ----- Prepara log-prezzo e rimuovi trend lineare (così lo ricostruiamo separatamente) -----
px = pd.Series(px).dropna()
px.index = pd.to_datetime(px.index)
lp = np.log(px).astype(float).copy()
t = np.arange(len(lp))
coef = np.polyfit(t, lp.values, 1)           # trend lineare: a*t + b
trend = np.polyval(coef, t)
resid = lp.values - trend                    # residuo (oscillazioni attorno al trend)

N = len(resid)
freqs = np.fft.rfftfreq(N, d=1.0)
F = np.fft.rfft(resid)                       # FFT 1D del residuo
power = np.abs(F)

# ----- Prendi TOP_K frequenze (escludi DC = indice 0) -----
order = np.argsort(power[1:])[::-1][:TOP_K] + 1
top_idx = np.sort(order)
top_freqs = freqs[top_idx]
top_amps  = (2.0 / N) * np.abs(F[top_idx])   # ampiezze reali (per k>0)
top_phase = np.angle(F[top_idx])             # fasi

# ----- Funzione di sintesi (passato + futuro) -----
def synth_path(T_future=60, amp_scale=1.0, phase_jitter=0.0, seed=None):
    rng = np.random.default_rng(seed)
    # tempi (passato + futuro)
    t2 = np.arange(N + T_future)
    # trend esteso
    trend2 = np.polyval(coef, t2)
    # residuo sintetico (somma di coseni)
    phases = top_phase.copy()
    if phase_jitter and phase_jitter > 0:
        phases = phases + rng.normal(0.0, phase_jitter, size=len(phases))
    resid2 = np.zeros_like(t2, dtype=float)
    for A, w, ph in zip(top_amps * amp_scale, 2*np.pi*top_freqs, phases):
        resid2 += A * np.cos(w * t2 + ph)
    lp2 = trend2 + resid2
    px2 = np.exp(lp2)

    # indice temporale passato + futuro con stessa frequenza del dato
    try:
        f = pd.infer_freq(px.index)
    except:
        f = None
    if f:
        future_idx = pd.date_range(px.index[-1], periods=T_future+1, freq=f)[1:]
    else:
        future_idx = pd.RangeIndex(start=len(px), stop=len(px)+T_future)
    idx2 = px.index.append(pd.Index(future_idx))
    return pd.Series(px2, index=idx2)

# ----- Costruisci scenari -----
scenarios = [synth_path(HORIZON, AMP_SCALE, PHASE_JITTER, seed=i) for i in range(N_SCENARIOS)]

# ----- Plot: storico + scenari con area FUTURO -----
fig, ax = plt.subplots(figsize=(10,4))
ax.plot(px.index, px.values, label="Prezzo reale", linewidth=1.4, zorder=3)

# area “futuro”
if HORIZON > 0:
    try:
        f = pd.infer_freq(px.index)
        future_start = px.index[-1]
        future_end   = scenarios[0].index[-1]
        ax.axvspan(future_start, future_end, alpha=0.12, color="tab:orange", label="Futuro (sintetico)")
    except Exception:
        pass

for i, s in enumerate(scenarios, 1):
    ax.plot(s.index, s.values, "--", linewidth=1.1, alpha=0.9, label=f"Scenario {i}")

loc = mdates.AutoDateLocator(); ax.xaxis.set_major_locator(loc); ax.xaxis.set_major_formatter(mdates.ConciseDateFormatter(loc))
ax.set_title(f"Scenari sintetici da frequenze TOP_K={TOP_K} | H={HORIZON} | jitter={PHASE_JITTER} | amp×{AMP_SCALE}")
ax.grid(True); ax.legend(ncols=2)
plt.show()

# ----- Tabella riassuntiva frequenze selezionate -----
periodi = (1/np.maximum(top_freqs, 1e-12)).astype(int)
pd.DataFrame({"freq": top_freqs, "periodo_barre": periodi, "ampiezza": top_amps, "fase_rad": top_phase})
```

```python
# @title
# =========================  BLOCCO MINIMA AZIONE + GRAFICI  =========================
# Requisiti: pandas, numpy, plotly, (opzionale) yfinance se vuoi auto-fetch
import pandas as pd, numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# ------------------- Parametri regolabili -------------------
ALPHA     = 1.2      # peso parte cinetica (lisciatura)
BETA      = 0.001      # peso parte potenziale (aderenza a F)
LP_SPAN   = 30       # fondamental smoothing (EWMA span in giorni)
SLOPE_WIN = 10       # finestra (giorni) per slope del percorso a minima azione
SAVE_HTML = False    # True per salvare gli html nel working dir

# ------------------- Input px (prezzo) ----------------------
# Caso A: px già esiste
if "px" in globals():
    _px = globals()["px"].copy()
# Caso B: px non esiste -> prova auto-fetch via yfinance se hai TICKER/INTERVAL/PERIOD
elif all(k in globals() for k in ["TICKER", "INTERVAL", "PERIOD"]):
    import yfinance as yf
    df = yf.download(TICKER, period=PERIOD, interval=INTERVAL, auto_adjust=True, progress=False)
    if df.empty:
        raise RuntimeError("Download yfinance vuoto. Definisci px manualmente.")
    _px = df["Close"].rename("px")
else:
    raise RuntimeError("Manca 'px'. Definisci una Series prezzo o specifica TICKER/INTERVAL/PERIOD.")

# Normalizza indice
if not isinstance(_px.index, pd.DatetimeIndex):
    _px.index = pd.to_datetime(_px.index)
_px = _px.dropna().sort_index()
_px.name = "px"

# ------------------- Fondamentali F (low-pass) ---------------
# EWMA morbido come proxy di 'fondamentali'
px_F = _px.ewm(span=LP_SPAN, adjust=False).mean()

# ------------------- Minima Azione (risoluzione tridiagonale) -------------------
# Minimizza: J = sum_t [ BETA*(x_t - F_t)^2 ] + sum_t [ ALPHA*(x_t - x_{t-1})^2 ]
# Sistema lineare:
#  t=0:       (B+ A)*x0 - A*x1 = B*F0
#  0<t<n-1:   -A*x_{t-1} + (B+2A)*x_t - A*x_{t+1} = B*F_t
#  t=n-1:     -A*x_{n-2} + (B+ A)*x_{n-1} = B*F_{n-1}
A = float(ALPHA)
B = float(BETA)

idx = _px.index
F = px_F.reindex(idx).astype(float).values
n = len(F)
if n < 3:
    raise RuntimeError("Serie troppo corta per calcolare la Minima Azione (n<3).")

# diagonali del sistema tridiagonale
lower = np.full(n-1, -A, dtype=float)
diag  = np.full(n,   B + 2*A, dtype=float); diag[0] = B + A; diag[-1] = B + A
upper = np.full(n-1, -A, dtype=float)
rhs   = B * F

# Thomas algorithm
def solve_tridiag(a, b, c, d):
    # a: subdiag (n-1), b: diag (n), c: superdiag (n-1), d: rhs (n)
    n = len(b)
    c_ = np.zeros(n-1, dtype=float)
    d_ = np.zeros(n, dtype=float)
    b_ = b.astype(float).copy()

    c_[0] = c[0] / b_[0]
    d_[0] = d[0] / b_[0]
    for i in range(1, n-1):
        denom = b_[i] - a[i-1]*c_[i-1]
        c_[i] = c[i] / denom
        d_[i] = (d[i] - a[i-1]*d_[i-1]) / denom
    d_[n-1] = (d[n-1] - a[n-2]*d_[n-2]) / (b_[n-1] - a[n-2]*c_[n-2])

    x = np.zeros(n, dtype=float)
    x[-1] = d_[n-1]
    for i in range(n-2, -1, -1):
        x[i] = d_[i] - c_[i]*x[i+1]
    return x

x_star_vals = solve_tridiag(lower, diag, upper, rhs)
px_star = pd.Series(x_star_vals, index=idx, name="px_star")

# ------------------- Densità di azione -------------------
dX = px_star.diff().fillna(0.0)
kin_density = 0.5 * A * (dX**2)                           # α·ΔX²/2
pot_density = 0.5 * B * ((px_star - px_F)**2)             # β·(X−F)²/2

# ------------------- Segnali per "trend futuri" -------------------
# 1) pendenza locale del percorso a minima azione (slope su SLOPE_WIN)
slope = px_star.diff(SLOPE_WIN) / SLOPE_WIN
slope = slope.rename(f"Slope({SLOPE_WIN})")

# 2) residuo prezzo - percorso (divergenza) e suo z-score
resid = (_px - px_star).rename("Residuo(px - x*)")
resid_std = resid.rolling(60, min_periods=20).std()
z_resid = (resid / resid_std).clip(-5, 5).rename("Z-Residuo (±5 cap)")

# 3) fascia di confidenza attorno a x* (±1σ residui storici recenti)
sigma = resid.rolling(60, min_periods=20).std().rename("sigma")
band_upper = (px_star + sigma).rename("x* + 1σ")
band_lower = (px_star - sigma).rename("x* - 1σ")

# ============================ GRAFICO 1 ============================
fig1 = make_subplots(rows=1, cols=1)
fig1.add_trace(go.Scatter(x=_px.index, y=_px.values, name="Prezzo", mode="lines", line=dict(width=2),
                          hovertemplate="Data=%{x|%Y-%m-%d}<br>Prezzo=%{y:.2f}<extra></extra>"))
fig1.add_trace(go.Scatter(x=px_star.index, y=px_star.values, name="Percorso Min-Azione", mode="lines",
                          line=dict(width=2, dash="dash"),
                          hovertemplate="Data=%{x|%Y-%m-%d}<br>x*=%{y:.2f}<extra></extra>"))
fig1.add_trace(go.Scatter(x=px_F.index, y=px_F.values, name="Fondamentali (EWMA)",
                          mode="lines", line=dict(width=1.5, dash="dot"),
                          hovertemplate="Data=%{x|%Y-%m-%d}<br>F=%{y:.2f}<extra></extra>"))
# bande ±1σ
fig1.add_trace(go.Scatter(x=band_upper.index, y=band_upper.values, name="x* + 1σ",
                          mode="lines", line=dict(width=1), hoverinfo="skip", opacity=0.5))
fig1.add_trace(go.Scatter(x=band_lower.index, y=band_lower.values, name="x* - 1σ",
                          mode="lines", line=dict(width=1), hoverinfo="skip", opacity=0.5, fill="tonexty"))

fig1.update_layout(
    title="Prezzo vs Percorso a Minima Azione + Fondamentali (con fascia ±1σ)",
    hovermode="x unified",
    legend=dict(orientation="h", y=1.02, x=0),
    margin=dict(l=40, r=10, t=60, b=40),
    xaxis=dict(type="date", rangeslider=dict(visible=True),
               rangeselector=dict(buttons=[
                   dict(count=1,  label="1M", step="month", stepmode="backward"),
                   dict(count=3,  label="3M", step="month", stepmode="backward"),
                   dict(count=6,  label="6M", step="month", stepmode="backward"),
                   dict(count=1,  label="1Y", step="year",  stepmode="backward"),
                   dict(step="all", label="All")
               ])),
    yaxis=dict(zeroline=False),
)
fig1.show()

# ============================ GRAFICO 2 ============================
dens_df = pd.DataFrame({
    "Cinetica (α·ΔX²/2)": kin_density,
    "Potenziale (β·(X−F)²/2)": pot_density
}).fillna(0.0)
dens_df["Totale"] = dens_df.sum(axis=1)
dens_df["Cumulata"] = dens_df["Totale"].cumsum()

fig2 = make_subplots(rows=2, cols=1, shared_xaxes=True, row_heights=[0.6, 0.4],
                     subplot_titles=("Densità di azione nel tempo", "Azione cumulata"))
fig2.add_trace(go.Scatter(x=dens_df.index, y=dens_df["Cinetica (α·ΔX²/2)"], name="Cinetica",
                          mode="lines", line=dict(width=1.6)), row=1, col=1)
fig2.add_trace(go.Scatter(x=dens_df.index, y=dens_df["Potenziale (β·(X−F)²/2)"], name="Potenziale",
                          mode="lines", line=dict(width=1.6)), row=1, col=1)
fig2.add_trace(go.Scatter(x=dens_df.index, y=dens_df["Cumulata"], name="Cumulata",
                          mode="lines", line=dict(width=2)), row=2, col=1)
fig2.update_layout(
    title="Azione: densità e cumulata",
    hovermode="x unified",
    legend=dict(orientation="h", y=1.02, x=0),
    margin=dict(l=40, r=10, t=60, b=40),
    xaxis2=dict(type="date", rangeslider=dict(visible=True),
                rangeselector=dict(buttons=[
                    dict(count=1,  label="1M", step="month", stepmode="backward"),
                    dict(count=3,  label="3M", step="month", stepmode="backward"),
                    dict(count=6,  label="6M",  step="month", stepmode="backward"),
                    dict(count=1,  label="1Y",  step="year",  stepmode="backward"),
                    dict(step="all", label="All")
                ])),
    yaxis=dict(zeroline=False), yaxis2=dict(zeroline=False)
)
fig2.show()

# ============================ GRAFICO 3 ============================
# Pannello segnali: slope (indicazione trend) + z-residuo (divergenza)
fig3 = make_subplots(rows=2, cols=1, shared_xaxes=True, row_heights=[0.5, 0.5],
                     subplot_titles=(f"Slope di x* su {SLOPE_WIN} giorni", "Z-Residuo (px - x*)"))
fig3.add_trace(go.Scatter(x=slope.index, y=slope.values, name=f"Slope({SLOPE_WIN})",
                          mode="lines", line=dict(width=1.6)), row=1, col=1)
fig3.add_hline(y=0, line=dict(width=1, dash="dot"), row=1, col=1)

fig3.add_trace(go.Scatter(x=z_resid.index, y=z_resid.values, name="Z-Residuo",
                          mode="lines", line=dict(width=1.6)), row=2, col=1)
fig3.add_hline(y=0, line=dict(width=1, dash="dot"), row=2, col=1)
fig3.add_hrect(y0=1.5, y1=3.0, line_width=0, fillcolor="orange", opacity=0.12, row=2, col=1)
fig3.add_hrect(y0=-3.0, y1=-1.5, line_width=0, fillcolor="orange", opacity=0.12, row=2, col=1)

fig3.update_layout(
    title="Indicatori: pendenza percorso a minima azione e divergenza normalizzata",
    hovermode="x unified",
    legend=dict(orientation="h", y=1.02, x=0),
    margin=dict(l=40, r=10, t=60, b=40),
    xaxis2=dict(type="date", rangeslider=dict(visible=True)),
)
fig3.show()

# ---------------------- Export opzionale ----------------------
if SAVE_HTML:
    import plotly.io as pio
    pio.write_html(fig1, "minima_azione_prezzo.html", auto_open=False)
    pio.write_html(fig2, "minima_azione_azione.html", auto_open=False)
    pio.write_html(fig3, "minima_azione_segnali.html", auto_open=False)
    print("Salvati: minima_azione_prezzo.html, minima_azione_azione.html, minima_azione_segnali.html")

# ---------------------- Variabili esposte ----------------------
# px_star, px_F, kin_density, pot_density, slope, z_resid, band_upper, band_lower
# ==================================================================================


```

```python
# ——— 14) Helper: z-score robusto, minima azione, estrattori di picchi ———
import numpy as np, pandas as pd, yfinance as yf, math, warnings, functools, itertools, gc, datetime as dt
from scipy import sparse
from numpy.linalg import solve

warnings.filterwarnings("ignore")

# Fallback parametri se non definiti in alto
try:    ALPHA
except: ALPHA = 5.0    # peso reg. smooth (seconda somma)
try:    BETA
except: BETA  = 1.0    # attacco al fondamentale
try:    LP_SPAN
except: LP_SPAN = 50   # EWMA span per fondamentale
try:    TOP_K
except: TOP_K = 6
try:    BANDWIDTH
except: BANDWIDTH = 1
try:    INTERVAL
except: INTERVAL = "1d"
try:    PERIOD
except: PERIOD = "5y"

def robust_z(x, win=252):
    """Z-score robusto su finestra: (x - mediana_rolling) / (1.4826 * MAD_rolling)."""
    s = pd.Series(x, dtype='float64')
    med = s.rolling(win, min_periods=max(20, win//5)).median()
    mad = (s.rolling(win, min_periods=max(20, win//5))
             .apply(lambda v: np.median(np.abs(v - np.median(v))), raw=True))
    z = (s - med) / (1.4826 * (mad.replace(0, np.nan)))
    return z.fillna(0.0).to_numpy()

def ewma(series, span):
    return pd.Series(series, dtype='float64').ewm(span=span, adjust=False).mean().to_numpy()

def minimize_action(px, lp_span=LP_SPAN, alpha=ALPHA, beta=BETA):
    """Ritorna x_star (percorso a minima azione), fondamentale F, densità cinetica/potenziale, slope, z_resid."""
    p = pd.Series(px, dtype='float64').dropna().to_numpy()
    if len(p) < 30:
        raise ValueError("Serie troppo corta")
    F = ewma(p, lp_span)

    n = len(p)
    # Costruisci matrice tridiagonale per termine (x_t - x_{t-1})^2 ~ L^T L
    main = (beta + 2*alpha) * np.ones(n)
    off  = -alpha * np.ones(n-1)
    # Correggi estremi (bordo)
    main[0]  = beta + alpha
    main[-1] = beta + alpha

    A = sparse.diags([off, main, off], [-1,0,1], shape=(n,n)).tocsc()
    b = beta * F
    x_star = sparse.linalg.spsolve(A, b)

    # Densità "energetiche"
    vel = np.diff(x_star, prepend=x_star[0])
    kin = vel**2                      # energia cinetica (densità)
    pot = (x_star - F)**2 * beta      # energia potenziale (densità)

    # pendenza locale e z-residuo
    slope = pd.Series(x_star).diff().ewm(span=10, adjust=False).mean().to_numpy()
    resid = pd.Series(p - x_star)
    z_res = (resid - resid.rolling(200, min_periods=30).mean()) / (resid.rolling(200, min_periods=30).std())
    z_res = z_res.fillna(0.0).to_numpy()

    return x_star, F, kin, pot, slope, z_res

def peak_summary(arr, index, recent_days=30, z_win=252):
    """Valuta picchi storici e recenti via z-score robusto + max locale recente."""
    s = pd.Series(arr, index=index)
    rz = pd.Series(robust_z(s.values, win=z_win), index=index)
    recent = s.loc[index[-recent_days:]] if len(index) >= recent_days else s
    rmax_val = recent.max()
    rmax_t   = recent.idxmax()
    return {
        "max_all": float(s.max()),
        "max_all_t": pd.to_datetime(s.idxmax()).date() if len(s)>0 else None,
        "recent_max": float(rmax_val) if len(recent)>0 else np.nan,
        "recent_max_t": pd.to_datetime(rmax_t).date() if len(recent)>0 else None,
        "recent_z_max": float(rz.loc[recent.index].max()) if len(recent)>0 else np.nan,
        "recent_z_at_max": float(rz.loc[rmax_t]) if len(recent)>0 else np.nan
    }
```

```python
# ——— 15) Batch scan robusto: scarica tutti i ticker, calcola x*, energie, misura picchi; salva CSV ———
from concurrent.futures import ThreadPoolExecutor, as_completed
import numpy as np
import pandas as pd
import yfinance as yf

CSV_PATH = "instruments.csv"
df_instr = pd.read_csv(CSV_PATH)
assert "ticker" in df_instr.columns, "Nel CSV deve esserci una colonna 'ticker'."

# Log semplice per capire perché un ticker fallisce
def fetch_series(ticker, period=PERIOD, interval=INTERVAL):
    try:
        df = yf.download(ticker, period=period, interval=interval, auto_adjust=True, progress=False)
        if df is None or df.empty or "Close" not in df:
            return ticker, None, "empty"
        s = df["Close"].dropna()
        return ticker, s, None
    except Exception as e:
        return ticker, None, str(e)

def analyze_one(row, period=PERIOD, interval=INTERVAL, recent_days=30, z_win=252, min_len=80):
    ticker = str(row["ticker"]).strip()
    name   = row.get("name", "")
    cat    = row.get("category", "")
    tk, s, err = fetch_series(ticker, period, interval)
    if s is None:
        return {"_status":"skip","ticker":ticker,"name":name,"category":cat,"reason":f"download:{err}"}
    if len(s) < max(min_len, recent_days+20):
        return {"_status":"skip","ticker":ticker,"name":name,"category":cat,"reason":f"too_short:{len(s)}"}
    try:
        x_star, F, kin, pot, slope, z_res = minimize_action(s.values)
        kin_peak = peak_summary(kin, s.index, recent_days, z_win)
        pot_peak = peak_summary(pot, s.index, recent_days, z_win)
        last_price = float(s.iloc[-1])
        last_date  = str(pd.to_datetime(s.index[-1]).date())
        last_slope = float(slope[-1])
        last_zres  = float(z_res[-1])
        score = np.nanmax([kin_peak["recent_z_max"], pot_peak["recent_z_max"]])
        return {
            "_status":         "ok",
            "ticker":          ticker,
            "name":            name,
            "category":        cat,
            "last_price":      last_price,
            "last_date":       last_date,
            "kin_recent_z_max": kin_peak["recent_z_max"],
            "kin_recent_max":   kin_peak["recent_max"],
            "kin_recent_max_t": kin_peak["recent_max_t"],
            "kin_max_all":      kin_peak["max_all"],
            "kin_max_all_t":    kin_peak["max_all_t"],
            "pot_recent_z_max": pot_peak["recent_z_max"],
            "pot_recent_max":   pot_peak["recent_max"],
            "pot_recent_max_t": pot_peak["recent_max_t"],
            "pot_max_all":      pot_peak["max_all"],
            "pot_max_all_t":    pot_peak["max_all_t"],
            "last_slope":       last_slope,
            "last_z_resid":     last_zres,
            "spike_score":      float(score) if np.isfinite(score) else np.nan,
        }
    except Exception as e:
        return {"_status":"skip","ticker":ticker,"name":name,"category":cat,"reason":f"min_action:{str(e)}"}

def run_batch_scan(category_filter=None, max_workers=4, recent_days=30, z_win=252, min_len=80):
    scan_list = df_instr.copy()
    if category_filter:
        scan_list = scan_list[scan_list["category"].astype(str).str.fullmatch(str(category_filter))]

    recs = []
    from concurrent.futures import ThreadPoolExecutor, as_completed
    with ThreadPoolExecutor(max_workers=max_workers) as ex:
        futs = [ex.submit(analyze_one, row, PERIOD, INTERVAL, recent_days, z_win, min_len)
                for _, row in scan_list.iterrows()]
        for f in as_completed(futs):
            recs.append(f.result())

    df_all = pd.DataFrame(recs)
    ok_df  = df_all[df_all["_status"]=="ok"].drop(columns=["_status"], errors="ignore")
    skip_df= df_all[df_all["_status"]=="skip"].drop(columns=["_status"], errors="ignore")

    if ok_df.empty:
        print("⚠️ Nessun ticker valido con questi parametri.")
        if not skip_df.empty:
            display(skip_df.head(10))
        cols = ["ticker","name","category","last_price","last_date",
                "kin_recent_z_max","pot_recent_z_max","last_slope","last_z_resid","spike_score"]
        return pd.DataFrame(columns=cols)

    if "spike_score" not in ok_df.columns:
        ok_df["spike_score"] = np.nan

    res = ok_df.sort_values("spike_score", ascending=False, na_position="last").reset_index(drop=True)
    res.to_csv("scan_results.csv", index=False)
    print(f"✅ OK: {len(res)}  | ❌ Skip: {len(skip_df)}  | Tot: {len(df_all)}")
    return res

# ESECUZIONE
batch_df = run_batch_scan(category_filter=None, max_workers=6, recent_days=30, z_win=252, min_len=80)
batch_df.head(20)
```

```python
# ——— 16) Filtri interattivi e link al singolo ticker ———
import plotly.express as px
import ipywidgets as w
from IPython.display import display, HTML

if 'batch_df' not in globals() or batch_df is None or batch_df.empty:
    print("Esegui prima la cella 15) per popolare 'batch_df'.")
else:
    z_slider = w.FloatSlider(value=2.0, min=0.0, max=6.0, step=0.1, description='min z-peak')
    rec_days = w.IntSlider(value=30, min=5, max=250, step=5, description='recent_days')
    sort_by  = w.Dropdown(options=[
        ('Score (max z)', 'spike_score'),
        ('Kin z-max (recent)', 'kin_recent_z_max'),
        ('Pot z-max (recent)', 'pot_recent_z_max'),
        ('|slope|', 'abs_slope'),
        ('|z_resid|', 'abs_zres')
    ], value='spike_score', description='Sort by')

    search = w.Text(value='', description='Search Ticker/Name', placeholder='es. NVDA, EURUSD, S&P...')
    top_n   = w.IntSlider(value=50, min=5, max=min(300, len(batch_df)), step=5, description='Top N')

    out_tbl = w.Output()
    out_fig = w.Output()

    # colonne ausiliarie
    batch_df['abs_slope'] = batch_df['last_slope'].abs()
    batch_df['abs_zres']  = batch_df['last_z_resid'].abs()

    def refresh(*_):
        with out_tbl:
            out_tbl.clear_output(wait=True)
            df = batch_df.copy()
            # filtro z
            df = df[(df['kin_recent_z_max']>=z_slider.value) | (df['pot_recent_z_max']>=z_slider.value)]
            # (ri)calcola recenti rispetto a rec_days? -> i valori sono calcolati con recent_days usati nello scan.
            # filtro testuale
            q = search.value.strip().upper()
            if q:
                df = df[df.apply(lambda r: q in str(r['ticker']).upper() or q in str(r['name']).upper(), axis=1)]
            # ordinamento
            df = df.sort_values(sort_by.value, ascending=False).head(top_n.value)
            # stampa tabella
            display(df[['ticker','name','category','last_date','last_price',
                        'kin_recent_z_max','pot_recent_z_max','spike_score',
                        'last_slope','last_z_resid']].style.format({
                'last_price': '{:,.2f}',
                'kin_recent_z_max': '{:.2f}',
                'pot_recent_z_max': '{:.2f}',
                'spike_score': '{:.2f}',
                'last_slope': '{:.4f}',
                'last_z_resid': '{:.2f}'
            }))
        with out_fig:
            out_fig.clear_output(wait=True)
            if df.empty:
                display(HTML("<i>Nessun ticker supera la soglia selezionata.</i>"))
                return
            fig = px.bar(df.head(top_n.value),
                         x='ticker',
                         y='spike_score',
                         hover_data=['name','kin_recent_z_max','pot_recent_z_max','last_price','last_date'],
                         title='Rilevanza picchi (max z di kin/pot)')
            fig.update_layout(xaxis_title="Ticker", yaxis_title="Spike score (max z)")
            display(fig)

    for wgt in (z_slider, rec_days, sort_by, search, top_n):
        wgt.observe(refresh, 'value')

    # Pulsante per impostare il ticker nel widget esistente (se presente)
    ticker_set = w.Text(description='Set TICKER →', placeholder='es. NVDA, SPY, BTC-USD')
    btn_set = w.Button(description='Invia al widget', button_style='info')
    lbl = w.HTML("<i>Se il notebook ha un widget <b>ticker_dd</b> o una variabile globale <b>TICKER</b>, li aggiorno.</i>")
    def _on_set(_):
        val = ticker_set.value.strip().upper()
        if not val:
            return
        try:
            # prova a impostare un dropdown chiamato ticker_dd
            if 'ticker_dd' in globals():
                try:
                    ticker_dd.value = val
                except Exception:
                    # se non presente tra le options, lo aggiunge
                    if hasattr(ticker_dd, 'options'):
                        opts = list(ticker_dd.options)
                        if val not in opts:
                            opts.append(val)
                            ticker_dd.options = opts
                        ticker_dd.value = val
            # aggiorna anche la variabile TICKER se esiste
            globals()['TICKER'] = val
            print(f"Ticker impostato: {val}. Ora riesegui i blocchi single-ticker.")
        except Exception as e:
            print("Non sono riuscito ad aggiornare il widget. Imposta manualmente il ticker:", val, "-", e)

    btn_set.on_click(_on_set)

    ui = w.VBox([
        w.HBox([z_slider, sort_by]),
        w.HBox([search, top_n]),
        w.HBox([ticker_set, btn_set]),
        lbl,
        out_fig,
        out_tbl
    ])
    display(ui)
    refresh()
```
